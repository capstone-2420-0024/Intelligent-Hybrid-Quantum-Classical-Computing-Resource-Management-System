```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def draw_system_architecture():
    fig, ax = plt.subplots(figsize=(12, 7))

    # 
    components = {
        "TaskSimulator": (0.5, 0.85),
        "Task DAG (with metadata)": (0.5, 0.72),
        "SchedulingEnv": (0.5, 0.58),
        "Observation": (0.25, 0.45),
        "Available Actions": (0.5, 0.45),
        "Reward": (0.75, 0.45),
        "Multi-Agent RL": (0.5, 0.3),
        "QMIX / VDN / PPO / LLM Agent": (0.5, 0.15)
    }

    for text, (x, y) in components.items():
        ax.text(x, y, text, ha='center', va='center', fontsize=10,
                bbox=dict(boxstyle="round,pad=0.3", edgecolor='gray', facecolor='lightyellow'))

    # 
    edges = [
        ("TaskSimulator", "Task DAG (with metadata)"),
        ("Task DAG (with metadata)", "SchedulingEnv"),
        ("SchedulingEnv", "Observation"),
        ("SchedulingEnv", "Available Actions"),
        ("SchedulingEnv", "Reward"),
        ("Observation", "Multi-Agent RL"),
        ("Available Actions", "Multi-Agent RL"),
        ("Reward", "Multi-Agent RL"),
        ("Multi-Agent RL", "QMIX / VDN / PPO / LLM Agent")
    ]

    for src, dst in edges:
        x1, y1 = components[src]
        x2, y2 = components[dst]
        ax.annotate("", xy=(x2, y2), xytext=(x1, y1),
                    arrowprops=dict(arrowstyle="->", color='black'))

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    plt.title("Task Scheduling Simulation and Reinforcement Learning Overall Architecture Diagram", fontsize=13)
    plt.show()

# 
draw_system_architecture()
```
